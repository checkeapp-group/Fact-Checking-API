# Run a Fact Checking 

This notebooks runs a single fact-checking using the full fact-checking Workflow from a configuration file

```python
import sys

sys.path.append("..")
```

```python
from dotenv import load_dotenv

from veridika.src.workflows import Workflow

load_dotenv(".env")
```

```python
workflow = Workflow(config_path="../configs/pipeline_configs/gemini_rag.yaml")

result, cost = await workflow.run(
    statement="Los coches eléctricos son más contaminantes que los coches de gasolina",
    internal_language="es",
    output_language="es",
    location="es",
    generate_image=True,
    do_rag=True,
)

print(f"📊 Workflow completed in {workflow.get_runtime()} seconds")
print(f"💰 Total cost: ${cost:.4f}")
print(f"📞 Total API calls: {result['logs']['summary']['total_calls']}")
print(f"🤖 Agents used: {len(workflow.get_agents())}")

# Cost per agent
print()
for agent in workflow.get_agents():
    print(
        f"🤖 {agent.name}\n  - cost: ${agent.cost:.4f}\n  - total run time: {agent.total_run_time:.2f} seconds"
    )


print(f"\n📝 Question: {result['question']}")
print(f"✅ Label: {result['metadata']['label']}")
print(
    f"🖼️ Image URL: {result['image']['url']}"
    if result["image"]["url"]
    else "No image generated"
)
print(f"📚 Sources used: {len(result['sources'])}")
print(f"❓ Related questions: {len(result['related_questions'])}")

print(result["answer"])
```
